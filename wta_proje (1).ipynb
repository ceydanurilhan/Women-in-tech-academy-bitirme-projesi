Şimdi arenalarımızın kapasitesine göre büyük mü küçük mu orta mı bir arena olduklarını bulacak bir fonksiyon yazalım

```python
def func_arena_buyuklugu(Arena_Kapasitesi):
  if Arena_Kapasitesi >= 20000.0:
    return "Büyük"
  elif Arena_Kapasitesi < 20000.0 and Arena_Kapasitesi >= 19000.0:
    return "Orta"
  else:
    return "Küçük"

teams["Arena_Büyüklüğü"] = teams.Arena_Kapasitesi.apply(func_arena_buyuklugu)
```
